```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("C:\\Users\\pssbh\\OneDrive\\Sem 5\\DWDM Pro\\new merged all3.xlsx")
```

```python
df.head()
```

```python
df=df.drop(['Name','Hallticket number','Coed'],axis=1)
```

```python
df.head()
```

```python
df['status'] = df['status'].replace({"Qualified": 1, "Not Qualified": 0})
df['Gender']=df['Gender'].replace({"Male":1,"Female":0})
df.head()
```

```python
df['status'] = pd.to_numeric(df['status'], errors='coerce')
df['Gender'] = pd.to_numeric(df['Gender'], errors='coerce')
df['Intermediate_marks'] = pd.to_numeric(df['Intermediate_marks'], errors='coerce')
df.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.hist(df[df['status'] == 1]['Intermediate_marks'], bins=10,edgecolor='black')  # Set the range using 'range'

# Set title and labels
ax1.set_title('Histogram of Intermediate Marks for Qualified Students')
ax1.set_xlabel('Intermediate Marks')
ax1.set_ylabel('Frequency')

ax2.hist(df[df['status'] == 0]['Intermediate_marks'], bins=10,edgecolor='black')  # Set the range using 'range'

# Set title and labels
ax2.set_title('Histogram of Intermediate Marks for Not Qualified Students')
ax2.set_xlabel('Intermediate Marks')
ax2.set_ylabel('Frequency')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
qualified_df = df[df['status'] == 1]
gender_counts = qualified_df['Gender'].value_counts()
ax1.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
ax1.set_title('Distribution of Qualified People by Gender')
not_qualified=df[df['status']==0]
gender_counts=not_qualified['Gender'].value_counts()
ax2.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
ax2.set_title('Distribution of not Qualified People by Gender')
labels = ['Male', 'Female']
plt.legend(title='Gender', labels=labels, loc='upper right')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
qualified_df = df[df['status'] == 1]
gender_counts = qualified_df['Eamcet year'].value_counts()
ax1.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
ax1.set_title('Distribution of Qualified People by Gender')
not_qualified_df = df[df['status'] == 0]
gender_counts = not_qualified_df['Eamcet year'].value_counts()
ax2.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
ax2.set_title('Distribution of Not Qualified People by Gender')
plt.show()
```

```python
qualified_count = df['status'].sum()
not_qualified_count = len(df) - qualified_count
counts = [qualified_count, not_qualified_count]
labels = ['Qualified', 'Not Qualified']
colors = ['skyblue', 'lightcoral']
plt.figure(figsize=(6,6))
plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Qualified and Not Qualified People')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.legend(title='status', labels=labels, loc='upper right')
# Display the pie chart
plt.show()
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
